```python
!pip install pandas matplotlib

# ADD YOUR CODE ON THE NEXT LINE
import pandas as pd
import matplotlib.pyplot as plt

# Load the stock market data from a CSV file
# You need to have a CSV file, for example: 'stock_data.csv'
# The CSV file should include columns like 'Date', 'Open', 'High', 'Low', 'Close', 'Volume'

dummy_data = {
    'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05'],
    'Open': [100, 105, 102, 98, 101],
    'High': [110, 115, 108, 105, 106],
    'Low': [95, 100, 98, 95, 98],
    'Close': [105, 110, 100, 100, 102],
    'Volume': [1000, 1500, 1200, 800, 900]
}

df = pd.DataFrame(dummy_data)

# Convert the 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Plotting the closing price
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Close Price')
plt.title('Stock Closing Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()

# Plotting the stock volume
plt.figure(figsize=(12, 6))
plt.bar(df.index, df['Volume'], label='Volume')
plt.title('Stock Volume')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.show()

# Simple moving average (SMA) for 50 days
df['SMA50'] = df['Close'].rolling(window=50).mean()

# Plotting the closing price with SMA50
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['SMA50'], label='50-Day SMA', linestyle='--')
plt.title('Stock Closing Prices with 50-Day SMA')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
# Create a dummy dataset for testing

```
